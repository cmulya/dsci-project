```python
#SHOWING ONLY THE FIRST 5 ROWS
import pandas as pd

df = pd.read_csv("/content/drive/My Drive/dataset/heart.csv")
df.head(5)


```

```python
#FILTERING DATA (40 < AGE < 50)
df2 = df[(df['age']>40)&(df['age']<50)]
df2
```

```python
#MACHINE LEARNING DESIGN TO PREDICT WHETHER A PATIENT HAS A HEART CONDITION
#(= 1) OR NO (=0) BASED ON THE 13 PARAMETERS
X = df.iloc[:, 0:13]
y = df.iloc[:, 13]
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.1)

```

```python
from sklearn.svm import SVC

clf = SVC()
clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)


```

```python
#THIS MODEL HAS AN ACCURACY OF 80%
from sklearn.metrics import accuracy_score

print(accuracy_score(y_pred, y_test))
```
